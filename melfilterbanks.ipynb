```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import librosa
import librosa.display
import IPython.display as ipd
from IPython.display import Audio
```

```python
suara = 'LonchuraLeucogastroide2.wav'
y , sr = librosa.load(suara, mono=True, sr=16000, duration=7)
mel = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=512, window='hann', n_mels=128, fmax=8000)
MFCCs13 = librosa.feature.mfcc(S=librosa.power_to_db(mel), sr=sr, appendEnergy=True, n_mfcc=13)
MFCCs26 = librosa.feature.mfcc(S=librosa.power_to_db(mel), sr=sr, appendEnergy=True, n_mfcc=26)
MFCCs26 = librosa.feature.delta(MFCCs26)
MFCCs39 = librosa.feature.mfcc(S=librosa.power_to_db(mel), sr=sr, appendEnergy=True, n_mfcc=39)
MFCCs39 = librosa.feature.delta(MFCCs39, order=2)
#MFCCs = librosa.feature.mfcc(y, sr)
Nmean =  np.mean(MFCCs13)
Nvar = np.var(MFCCs13)
Nmax = np.amax(MFCCs13)
Nmin = np.amin(MFCCs13)

print('mel : ',mel.shape)
print('mfcc13 : ',MFCCs13.shape, 'mean :', Nmean, 'variance :', Nvar, ' max :', Nmax, ' min :', Nmin )
print('mfcc26 : ',MFCCs26.shape)
print('mfcc39 : ',MFCCs39.shape)
#graph
plt.title('mfcc graph')
librosa.display.specshow(MFCCs13, sr = sr, x_axis = 'time')

plt.figure()
librosa.display.specshow(mel, x_axis='linear')
plt.ylabel('Mel filter')
plt.title('Mel filter bank')
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
import os
import csv
# generating a dataset jalanin kalaw masih belom poenya data.csv
header = 'filename mfcc13mean mfcc13var mfcc13max mfcc13min mfcc26mean mfcc26var mfcc26vmax mfcc26min mfcc39mean mfcc39var mfcc39vmax mfcc39min label'
#header += ' label'
header = header.split()

file = open('databurung.csv', 'w', newline='')
with file:
    writer = csv.writer(file)
    writer.writerow(header)
genres = 'blues classical country disco hiphop jazz metal pop reggae rock'.split()
for g in genres:
    for filename in os.listdir(f'./genres/{g}'):
        namaswra = f'./genres/{g}/{filename}'
        y , sr = librosa.load(namaswra, sr=16000, mono=True, duration=5)
        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=1024, hop_length=512, window='hann', n_mels=128, fmax=8000)
        MFCCs13 = librosa.feature.mfcc(S=librosa.power_to_db(mel), sr=sr, dct_type=2,  appendEnergy=True, n_mfcc=13)
        MFCCs26 = librosa.feature.mfcc(S=librosa.power_to_db(mel), sr=sr, dct_type=2,  appendEnergy=True, n_mfcc=26)
        MFCCs26 = librosa.feature.delta(MFCCs26)
        MFCCs39 = librosa.feature.mfcc(S=librosa.power_to_db(mel), sr=sr, dct_type=2,  appendEnergy=True, n_mfcc=39)
        MFCCs39 = librosa.feature.delta(MFCCs39, order=2)
        #mengambil nilai mean saja
        to_append = f'{filename} {np.mean(MFCCs13)} {np.var(MFCCs13)} {np.amax(MFCCs13)} {np.amin(MFCCs13)} {np.mean(MFCCs26)} {np.var(MFCCs26)} {np.amax(MFCCs26)} {np.amin(MFCCs26)} {np.mean(MFCCs39)} {np.var(MFCCs39)} {np.amax(MFCCs39)} {np.amin(MFCCs39)} {g}'
        file = open('databurung.csv', 'a', newline='')
        with file:
            writer = csv.writer(file)
            writer.writerow(to_append.split())
```
